```python
# If the size of your data is managable, you can directly load the csv file as a DataFrame using Pandas:
import pandas as pd
# data: https://data.cambridgema.gov/Transportation-Planning/Resident-Workforce-Commuting-Mode-Split-by-Place-o/4wd7-ic9p
url = 'https://data.cambridgema.gov/resource/4wd7-ic9p.csv'
modes = pd.read_csv(url)
modes.head()

```

```python
for i in modes.index:
    modes.loc[i,'top_mode'] = modes.iloc[:, 6:].loc[i].sort_values(ascending=False).index[0]
```

```python
modes
```

```python
modes_1990 = modes[modes.year =='1990']
modes_2000 = modes[modes.year =='2000']
modes_2006 = modes[modes.year =='2006-2010']
modes_2012 = modes[modes.year =='2012-2016']
```

```python

modes_1990.to_csv('modes_1990.csv')
modes_2000.to_csv('modes_2000.csv')
modes_2006.to_csv('modes_2006.csv')
modes_2012.to_csv('modes_2012.csv')
```
